```python
# First, we need to import some libraries to help us with our work
import pandas as pd  # This helps us organize and work with data (like tables)
import seaborn as sns  # This helps us create beautiful graphs
import matplotlib.pyplot as plt  # This helps us draw our graphs
import numpy as np  # This helps with mathematical operations

# Let's set a nicer style for our graphs
sns.set(style="whitegrid")

# Now, we'll load the Titanic dataset that comes built-in with seaborn
titanic_data = sns.load_dataset('titanic')

# Let's take a look at the first few rows of the dataset to see what it looks like
print("Here’s a peek at the Titanic data:")
print(titanic_data.head())

# Next, we'll clean up the data by fixing any missing pieces
def clean_data(data):
    # If someone's age is missing, we'll replace it with the average age
    data['age'] = data['age'].fillna(data['age'].mean())
    # If we don’t know where someone boarded the Titanic, we will use the most common boarding point
    data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
    # Remove any duplicate rows from the dataset
    data.drop_duplicates(inplace=True)
    return data

# Clean the Titanic data and save the cleaned version into a new variable
cleaned_data = clean_data(titanic_data)

# Function to show the distribution of passenger ages
def plot_age_distribution(df):
    plt.figure(figsize=(10, 6))  # Make the graph a bit bigger
    sns.boxplot(x=df['age'], color='skyblue')  # Create a box plot to show the age distribution
    plt.title('How Old Were the Titanic Passengers?', fontsize=16, color='navy')  # Title of the graph
    plt.xlabel('Age', fontsize=12)  # Label for the x-axis
    plt.grid(True)  # Add gridlines for easier reading
    plt.show()  # Display the graph

# Function to show how many people survived
def plot_survivor_counts(df):
    plt.figure(figsize=(10, 6))  # Set the size of the graph
    sns.countplot(x='survived', data=df, palette='pastel')  # Count how many survived or didn’t
    plt.title('How Many People Survived?', fontsize=16, color='darkred')  # Title of the graph
    plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)  # Label for the x-axis
    plt.ylabel('Number of People', fontsize=12)  # Label for the y-axis
    plt.grid(True)  # Add gridlines for better visibility
    plt.show()  # Display the graph

# Function to show a histogram of passenger ages
def plot_age_histogram(df):
    plt.figure(figsize=(10, 6))  # Set the size of the graph
    df['age'].hist(bins=30, color='orange', edgecolor='black')  # Create a histogram for ages
    plt.title('How Many People Were in Each Age Group?', fontsize=16, color='darkgreen')  # Title of the graph
    plt.xlabel('Age', fontsize=12)  # Label for the x-axis
    plt.ylabel('Number of People', fontsize=12)  # Label for the y-axis
    plt.grid(axis='y')  # Add gridlines on the y-axis for easier reading
    plt.show()  # Display the graph

# Function to show a heatmap of correlations between numerical values
def plot_correlation_heatmap(df):
    # Select only the numeric columns for this graph
    numeric_df = df.select_dtypes(include=[np.number])
    plt.figure(figsize=(10, 6))  # Set the size of the graph
    sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis', fmt=".2f", linewidths=.5)  # Create a heatmap
    plt.title('How Are the Numbers Related?', fontsize=16, color='darkblue')  # Title of the graph
    plt.show()  # Display the graph

# Convert categorical variables (like 'male' and 'female') to numbers for analysis
cleaned_data['sex'] = cleaned_data['sex'].map({'male': 0, 'female': 1})  # 'male' becomes 0 and 'female' becomes 1
cleaned_data['embarked'] = cleaned_data['embarked'].map({'C': 0, 'Q': 1, 'S': 2})  # 'C', 'Q', 'S' become 0, 1, 2

# Now, let's create our graphs to visualize the data
plot_age_distribution(cleaned_data)  # Show the age distribution graph
plot_survivor_counts(cleaned_data)  # Show how many survived
plot_age_histogram(cleaned_data)  # Show the age groups histogram
plot_correlation_heatmap(cleaned_data)  # Show the correlation heatmap

# Let's share some interesting facts we found from the data
print("Here are some interesting facts we discovered:")
print(f"1. The average age of passengers is about {cleaned_data['age'].mean():.2f} years.")
print(f"2. A total of {cleaned_data['survived'].sum()} people survived out of {cleaned_data.shape[0]} passengers.")
print(f"3. Most passengers boarded from {cleaned_data['embarked'].mode()[0]} (which is coded as {cleaned_data['embarked'].mode()[0]}).")
```
